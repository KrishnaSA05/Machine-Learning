```python
import seaborn as sns
```

```python
sns.load_dataset('iris')
```

```python
df = sns.load_dataset('iris')
df.head()	
```

```python
#independent features and dependent features
X = df.iloc[:,:-1]  #independent features    
y = iris.target   #dependent feature
print(X)
print(y)
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
#Post Prunning Decision Tree Classifier
#treemodel = DecisionTreeClassifier() #takes default parameters from sklearn DecisionTreeClassifier module like criterion='gini', splitter='best' etc.
treemodel = DecisionTreeClassifier(max_depth=2)
```

```python
treemodel.fit(X_train, y_train) #fitted the model on training data
```

```python
#constructing decision tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treemodel, filled=True)
```

```python
#predictions
y_pred=treemodel.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
score = accuracy_score(y_test, y_pred)
print("Accuracy:", score)
```

```python
print(classification_report(y_test, y_pred))
```
