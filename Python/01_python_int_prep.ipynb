```python
# Q1: Check Even or Odd

def check_even_odd(num):
    if num % 2 == 0:
        print("Even")
    else:
        print("Odd")

num = int(input("Enter the number: "))
check_even_odd(num)
```

```python
#Q2: Check Prime Number

def is_prime(num):
    if num <= 1:
        return "Not Prime"
    for i in range(2, num):
        if num % i == 0:
            return "Not Prime"
    return "Prime"

num = int(input("Enter a number: "))
print(is_prime(num))
```

```python
#Q3: Factorial of a Number

def factorial(num):
    
    # Base case
    if num < 0:
        return "Invalid Input"
    
    result = 1
    for i in range(1, num + 1):
        result *= i
    return result

num = int(input("Enter a number: "))
print(factorial(num))
```

```python
#Q5: Reverse a Number

def reverse_number(num):
    reversed_num = 0

    while num > 0:
        digit = num % 10
        reversed_num = reversed_num * 10 + digit
        num = num // 10
    return reversed_num 

num = int(input("Enter a number: "))
print(reverse_number(num))
```

```python
#using string slicing
def reverse_number(num):
    return int(str(num)[::-1])

num = int(input("Enter a number: "))
print(reverse_number(num))
```

```python
# Q6: Check Palindrome Number

def is_palindrome(num):

    if num<0:
        return "Invalid Input"

    original = num
    reversed_num = 0

    while num>0:
        digit = num % 10
        reversed_num = reversed_num * 10 + digit
        num = num // 10

    # Can be written like this
    #return "Palindrome" if original == reversed_num else "Not Palindrome"

    if original == reversed_num:
        return "Palindrome"
    else :
        return "Not Palindrome"

num = int(input("Enter a number: "))
print(is_palindrome(num))
```

```python
# Q7. Armstrong Number

def is_armstrong(num):
    if num < 0:
        return "Not Armstrong"

    original = num
    total = 0
    power = len(str(num))      # dynamic digit count

    while num > 0:
        digit = num % 10
        total = total + digit ** power
        num = num // 10

    if original == total:
        print("Armstrong")
    else:
        print("Not Armstrong")

num = int(input("Enter a number: "))
is_armstrong(num)
```

```python
# Q8: Sum of Digits

def sum_of_digits(num):
    num = abs(num)

    total = 0

    while num > 0:
        digit = num % 10
        total = total + digit
        num = num // 10
    return total

num = int(input("Enter a number: "))
print(sum_of_digits(num))
```

```python
# Q9: Largest of Three Numbers

def largest_of_three(num1, num2, num3):
    if num1 >= num2 and num1 >= num3:
        return num1
    elif num2 >= num1 and num2 >= num3:
        return num2
    else:
        return num3


num1 = int(input("Enter a first number: "))
num2 = int(input("Enter a second number: "))
num3 = int(input("Enter a third number: "))

print(largest_of_three(num1, num2, num3))
```

```python
#Q10: GCD of Two Numbers
def find_gcd(a,b):
    a = abs(a)
    b = abs(b)

    while b != 0:
        a, b = b, a % b

    return a

a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))
print(find_gcd(a, b))
```

```python
# Q11. LCM of Two Numbers
# LCM= a * b // GCD

def find_gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

def find_lcm(a, b):
    a, b = abs(a), abs(b)
    if a == 0 or b == 0:   # Edge case
        return 0
    return (a * b) // find_gcd(a, b)

a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))
print(find_lcm(a, b))
```

```python
 #Q12: Check Leap Year

 def is_leap_year(year):
     if year <= 0:
         return "Invalid Year"

     if year % 400 == 0:
         return "Leap Year"
     elif year % 100 == 0:
         return "Not Leap Year"
     elif year % 4 == 0:
         return "Leap Year"
     else:
         return "Not Leap Year"

year = int(input("Enter a year: "))
print(is_leap_year(year))
```

```python
# Q13: Count Vowels and consonants

def count_vowels_consonants(s):
    if not s:
        return "Invalid Input"

    s = s.lower()
    vowels= "aeiou"
    vowels_count = 0
    consonants_count = 0

    for ch in s:
        if ch.isalpha():  #to check whether a string consists of only alphabetical characters
            if ch in vowels:
                vowels_count += 1
            else:
                consonants_count += 1
    print(f"Vowels: {vowels_count}")
    print(f"Consonants: {consonants_count}")

s=input("Enter the text: ")
count_vowels_consonants(s)
```

```python
# Q14: Reverse a String

def reverse_string(s):
    if not s:
        return "Invalid Input"

    reversed_str = ""
    for ch in s:
        reversed_str = ch + reversed_str
    return reversed_str

s=input("Enter the text: ")
print(reverse_string(s))
```

```python
# Q15: Check Anagram

def is_anagram(s1, s2):
    if not s1 or not s2:
        return "Invalid String"

    s1 = s1.lower().replace(" ", "")
    s2 = s2.lower().replace(" ", "")

    if len(s1) != len(s2):
        return "Not Anagram"

    freq= {}

    for ch in s1:
        freq[ch]=freq.get(ch, 0) + 13

    for ch in s2:
        if ch not in freq:
            return "Not Anagram"
        freq[ch] -= 1

    return "Anagram"

s1 = input("Enter first string: ")
s2 = input("Enter second string: ")
print(is_anagram(s1, s2))
```
