```python
# Q1: Check Even or Odd

def check_even_odd(num):
    if num % 2 == 0:
        print("Even")
    else:
        print("Odd")

num = int(input("Enter the number: "))
check_even_odd(num)
```

```python
#Q2: Check Prime Number

def is_prime(num):
    if num <= 1:
        return "Not Prime"
    for i in range(2, num):
        if num % i == 0:
            return "Not Prime"
    return "Prime"

num = int(input("Enter a number: "))
print(is_prime(num))
```
