```python
#Armstrong Number:
def armstrongCheck(num):
    first_num = num
    second_num = 0

    while num > 0:
        digit = num % 10
        second_num = second_num + (digit ** 3)
        num = num // 10

    if first_num == second_num:
        print("The number is Armstrong Number")
    else:
        print("The number is not Armstrong Number")

num = int(input("Enter the number: "))
armstrongCheck(num)
```

```python
#Problem — Find Missing Number
def missingNumber(numbers):
    actual_sum = 0
    expected_sum = 0

    for num in numbers:
        actual_sum += num

    for i in range(1, max(numbers) + 1):
        expected_sum += i

    missing = expected_sum - actual_sum

    return missing


numbers = [1,2,4,5,6]

print(missingNumber(numbers))
```

```python
# Move All Zeros to the End #Two pointer appraoch
def moveZeros(nums):
    l = 0
    for r in range(len(nums)):
        if nums[r] != 0:
            nums[l], nums[r] = nums[r], nums[l]
            l += 1
    return nums
nums = [0,1,0,3,12]
print(moveZeros(nums))
```

```python
#Simpler appraoch
def moveZeros(nums):
    result=[]
    zero_count = 0

    for num in nums:
        if num == 0:
            zero_count += 1
        else:
            result.append(num)
    result.extend([0] * zero_count)
    return result

nums = [0,1,0,3,12]
print(moveZeros(nums))
```

```python
# Problem 5 — Second Smallest Number

def secondSmallest(nums):
    smallest=nums[0]
    second_smallest=nums[0]

    for num in nums:

        if num < smallest:
            second_smallest = smallest
            smallest = num

        elif num < second_smallest and num != smallest:
            second_smallest = num

    return second_smallest

nums=[12, 4, 18, 2, 9]
secondSmallest(nums)
```

```python
# Longest Increasing Streak
def longestIncreasingStreak(numbers):
    current_streak = 1
    max_streak = 1

    for i in range(1, len(numbers)):

        if numbers[i] > numbers[i-1]:
            current_streak += 1

            if current_streak > max_streak:
                max_streak = current_streak

        else:
            current_streak = 1

    return max_streak

numbers = [1,2,3,1,2,5,6]

print(longestIncreasingStreak(numbers))
```

```python
def binarySearch1(nums, target):
    left=0
    right=len(nums)-1

    while left <= right:
        mid = (left + right)//2

        if nums[mid] == target:
            return mid
        
        elif nums[mid] < target:
            left = mid + 1

        else:
            right = mid - 1
    
    return -1

nums= [2, 5, 8, 12, 16, 23, 38]
target=12
print(binarySearch1(nums,target))
```

```python
def binarySearch2(nums, target):
    left=0
    right=len(nums)-1

    while left <= right:
        mid = (left + right)//2

        if nums[mid] == target:
            return True
        
        elif nums[mid] < target:
            left = mid + 1

        else:
            right = mid - 1
    
    return False

nums= [2, 5, 8, 12, 16, 23, 38]
target=12
print(binarySearch2(nums,target))
```

```python
# Find first occurences.
def firstOccurrence(nums, target):
    left = 0
    right = len(nums) - 1

    answer = -1

    while left <= right:

        mid = (left + right) // 2

        if nums[mid] == target:
            answer = mid
            right = mid - 1

        elif nums[mid] < target:
            left = mid + 1

        else:
            right = mid - 1

    return answer
```

```python
# Last occurence
def lastOccurrence(nums,target):

    left=0
    right=len(nums) - 1
    answer= -1

    while left <= right:

        mid=(left+right)//2
        
        if nums[mid] == target:
            answer = mid
            left = mid + 1

        elif nums[mid] < target:
            left = mid + 1

        else:
            right = mid - 1

    return answer

nums = [1,2,2,2,3,4,5]
target = 2
print(lastOccurrence(nums, target))
```

```python
# Count Occurrences

def countOccurrences(nums, target):

    first = firstOccurrence(nums, target)

    if first == -1:
        return 0

    last = lastOccurrence(nums, target)

    return last - first + 1

nums = [1,2,2,2,3,4,5]

print(countOccurrences(nums, 2))
```

```python
#Two Sum
def twoSum(nums, target):
    left = 0
    right = len(nums) - 1
    

    while left < right:
        current_sum = nums[left] + nums[right]
        if current_sum == target:
            return (nums[left], nums[right])

        elif current_sum<target:
            left += 1

        else:
            right -= 1
    return None

nums = [2, 5, 7, 11, 15]
target = 9
print(twoSum(nums, target))
```

```python
def two_sum(nums, target):
    hashmap = {}

    for i, num in enumerate(nums):
        complement = target - num

        if complement in hashmap:
            return [hashmap[complement], i]

        hashmap[num] = i

    return []

nums = [2, 5, 7, 11, 15]
target = 9
print(twoSum(nums, target))
```

```python
def buySellStock(prices):

    left = 0      # buy
    right = 1     # sell
    max_profit = 0
    while right < len(prices):

        if prices[right] > prices[left]:
            profit = prices[right] - prices[left]

            if profit > max_profit:
                max_profit = profit
        else:
            left = right
        right += 1
    return max_profit

prices = [7,1,5,3,6,4]
print(buySellStock(prices))
```

```python
# Container With Most Water

def containerMostWater(height):

    left = 0
    right = len(height) - 1

    max_area = 0

    while left < right:

        width = right - left

        current_height = min(height[left], height[right])

        area = width * current_height

        if area > max_area:
            max_area = area

        if height[left] < height[right]:
            left += 1
        else:
            right -= 1

    return max_area

height = [1,8,6,2,5,4,8,3,7]
print(containerMostWater(height))
```

```python
# Remove Duplicates from Sorted Array

def removeDuplicates(nums):

    if not nums:
        return 0

    left = 0

    for right in range(1, len(nums)):

        if nums[right] != nums[left]:

            left += 1
            nums[left] = nums[right]

    return left + 1

nums = [1,1,2,2,3,4,4]
print(removeDuplicates(nums))
```
