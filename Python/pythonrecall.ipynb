```python
def even_list(numbers):
    even_num_list = []

    for num in numbers:
        if num % 2 == 0:
            even_num_list.append(num)
    return even_num_list

numbers = [1, 2, 3, 4, 5, 6, 7, 8]
print(even_list(numbers))
```

```python
def countVowels(text):
    vowels_count = 0
    lowercase_vowels = 0
    uppercase_vowels = 0

    for letter in text:
        if letter in "aeiouAEIOU":
            vowels_count += 1

        if letter in "aeiou":
            lowercase_vowels += 1

        if letter in "AEIOU":
            uppercase_vowels += 1

    return vowels_count, lowercase_vowels, uppercase_vowels
text = "PyThOn AE"
print(countVowels(text))
```

```python
def word_count(sentence):
    words=sentence.split()
    word_count={}
    for word in words:
        word_count[word]=word_count.get(word, 0) + 1
    print(word_count)
    

sentence = "python is fun and python is easy"
word_count(sentence)
```

```python
numbers = [1,2,3,2,4,5,1,6]
seen = set()
duplicates = set()
for num in numbers:
    if num in seen:
        duplicates.add(num)
    else:
        seen.add(num)
print(list(duplicates))
```

```python
def duplicateNumbers(numbers):
    num_counter={}
    dup_numbers=[]

    for num in numbers:
        num_counter[num]=num_counter.get(num, 0) + 1

    for key,value in num_counter.items():
        if value > 1:
            dup_numbers.append(key)
    return dup_numbers

numbers = [1,2,3,2,4,5,1,6]
print(duplicateNumbers(numbers))
```
