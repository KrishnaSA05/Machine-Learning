```python
# Two Sum (Hash Map — O(n)) 
def twoSum(nums, target):
    seen = {} #stores num,index pair

    for i, n in enumerate(nums):
        diff = target - n

        if diff in seen:
            return [seen[diff], i]

        seen[n] = i #stores a key–value pair in the hashmap


nums = [2, 7, 11, 15]
target = 9

print(twoSum(nums, target))
```

```python
# Two Sum (Brute Force (for comparison, O(n²)))

def twoSum(nums, target):
    for i in range(len(nums)):
        for j in range(i + 1, len(nums)): # pick first number
            if nums[i] + nums[j] == target: #pick second number
                return [i, j]
nums = [2, 7, 11, 15]
target = 9

print(twoSum(nums, target))


#Brute force = “Try everything until something works”
#Hashmap = Store & look up fast
```

```python
#Prime Number:
#Eg: 2,3,5,7,11,13

num=int(input("Enter the number: "))
print(num)

count = 0

for i in range(1,num + 1):
    if num%i == 0:
        count += 1

if count == 2:
    print("Number is a prime number")
else:
    print("Number is not a prime number")

```

```python
#Palindrome → “same when reversed”
num= int(input("Enter the number: "))

original = num
reverse = 0

while num>0:
    digit = num % 10
    reverse = reverse*10 + digit
    num = num // 10

if original == reverse:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
def armstrongNum(num):
    n = len(str(num))
    real = num
    total = 0

    while num > 0:
        digit = num % 10
        total += digit ** n
        num = num // 10

    if total == real:
        print("Armstrong Number")
    else:
        print("Not Armstrong Number")


num = int(input("Enter the number: "))
armstrongNum(num)
```

```python
#Reverse the number:

def reverseNumber(num):

    reverse=0

    while num>0:
        digit = num % 10
        reverse = reverse * 10 + digit
        num = num // 10
    print(reverse)
    
num=int(input("Enter the number: "))
reverseNumber(num)
```

```python
#factorial of a number

def factorial(num):

    result=1

    for i in range(1, num+1):
        result = result * i

    print(result)

num = int(input("Enter a number: "))
factorial(num)
```
