```python
# Two Sum (Hash Map — O(n)) 
def twoSum(nums, target):
    seen = {} #stores num,index pair

    for i, n in enumerate(nums):
        diff = target - n

        if diff in seen:
            return [seen[diff], i]

        seen[n] = i #stores a key–value pair in the hashmap


nums = [2, 7, 11, 15]
target = 9

print(twoSum(nums, target))
```

```python
# Two Sum (Brute Force (for comparison, O(n²)))

def twoSum(nums, target):
    for i in range(len(nums)):
        for j in range(i + 1, len(nums)): # pick first number
            if nums[i] + nums[j] == target: #pick second number
                return [i, j]
nums = [2, 7, 11, 15]
target = 9

print(twoSum(nums, target))


#Brute force = “Try everything until something works”
#Hashmap = Store & look up fast
```

```python
#Palindrome → “same when reversed”
num= int(input("Enter the number: "))

original = num
reverse = 0

while num>0:
    digit = num % 10
    reverse = reverse*10 + digit
    num = num // 10

if original == reverse:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
def armstrongNum(num):
    n = len(str(num))
    real = num
    total = 0

    while num > 0:
        digit = num % 10
        total += digit ** n
        num = num // 10

    if total == real:
        print("Armstrong Number")
    else:
        print("Not Armstrong Number")


num = int(input("Enter the number: "))
armstrongNum(num)
```

```python
# Vowels count
def vowelsCount(sentence):
    count = 0

    for i in sentence:
        if i in "aeiouAEIOU":
            count += 1

    return count

sentence = "I love Python programming"
vowelsCount(sentence)
```

```python
# Move Zeroes
def moveZeroes(nums):
    l=0
    r=len(nums)
    for r in range(r):
        if nums[r] != 0:
            nums[l] , nums[r] = nums[r] , nums[l]
            l += 1 
    return nums      


nums= [0,1,0,3,12]
print(moveZeroes(nums))
```

```python
#Reverse the array

def reverseArray(arr):
    reverse=[]

    for num in range(len(arr)-1,-1,-1):
        reverse.append(arr[num])
    return reverse

arr = [1, 2, 3, 4]
reverseArray(arr)

#Using index
def reverseArray(arr):
    return arr[::-1]
arr = [1, 2, 3, 4]
reverseArray(arr)

#two pointer
def reverseArray(arr):
    left=0
    right=len(arr) -1 

    while left<right:
        arr[left], arr[right] = arr[right], arr[left]
        left +=1 
        right -=1
    return arr
arr = [1, 2, 3, 4]
reverseArray(arr)
```

```python
#Count even and odd numbers
def countEvenOdd(arr):
    even_count=0
    odd_count=0

    for num in arr:
        if num%2==0:
            even_count += 1
        else:
            odd_count +=1
    return {"even": even_count,"odd": odd_count}
    
arr = [1, 2, 3, 4, 5]
countEvenOdd(arr)
```

```python
#Find duplicate elements
def duplicateElements(arr):
    freq={}

    for num in arr:
        freq[num]=freq.get(num,0)+1
    
    duplicates=[]
    for key in freq:
        if freq[key]>1:
            duplicates.append(key)
    return duplicates
arr = [1, 2, 3, 2, 4, 1]
duplicateElements(arr)
```
