```python
# With Encapsulation:
class BankAccount:
    def __init__(self, account_holder, balance):
        self.account_holder = account_holder
        self._balance = balance

    def deposit(self, amount):
        if amount > 0:
            self._balance += amount

    def withdraw(self, amount):
        if amount > self._balance:
            print("Insufficient funds")
        else:
            self._balance -= amount

    def get_balance(self):
        return self._balance
    
# The single underscore (_) is a convention that means:
#"This attribute is intended for internal use. Please don't access it directly."
# Encapsulation is the practice of hiding an object's internal state and allowing access to it only through well-defined methods. It helps protect data integrity and reduces unintended modifications.
# the data and the methods operating on that data are grouped inside the BankAccount class. However, the internal state is not fully hidden because balance can still be modified directly from outside the class. A more complete implementation would make the balance private (or conventionally private in Python) and expose controlled access through methods.
    
```

```python
# Inheritance
class BankAccount:
    def __init__(self, account_holder, balance):
        self.account_holder = account_holder
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient funds")
        else:
            self.balance -= amount


class SavingsAccount(BankAccount):
    def add_interest(self):
        self.balance += self.balance * 0.05


class CurrentAccount(BankAccount):
    def withdraw(self, amount):
        # Allow overdraft up to -500
        if self.balance - amount < -500:
            print("Overdraft limit exceeded")
        else:
            self.balance -= amount


# Testing SavingsAccount
savings = SavingsAccount("Alice", 1000)
savings.add_interest()
print(f"Savings Balance: {savings.balance}")

# Testing CurrentAccount
current = CurrentAccount("Bob", 1000)

current.withdraw(1200)
print(f"Current Balance: {current.balance}")

current.withdraw(2000)

# Both SavingsAccount and CurrentAccount share common properties and behaviors such as account_holder, balance, deposit, and withdraw. Instead of duplicating code, we place the common functionality in BankAccount and let the specialized accounts inherit from it. This improves reusability and maintainability.
```

```python
# Polymorphism allows me to write code against a common type and let each object decide how to perform the operation.
class Dog:
    def speak(self):
        return "Bark"

class Cat:
    def speak(self):
        return "Meow"
    
animals = [Dog(), Cat()]

for animal in animals:
    print(animal.speak())

# The polymorphism is that this single line: animal.speak()
```
