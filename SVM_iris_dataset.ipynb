```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
%matplotlib inline
```

```python
#load the iris dataset
iris = load_iris() 
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df.head()
```

```python
df['target']= iris.target
df.head()
```

```python
iris.target_names
```

```python
df[df.target==1].head() #setosa, versicolor, virginica= 50 eachSS
```

```python
df['flower_name']=df.target.apply(lambda x: iris.target_names[x]) #map target numbers to flower names
df.head()
```

```python
#to separate different classes
df0=df[df.target==0] 
df1=df[df.target==1]
df2=df[df.target==2]
```

```python
df0.head()
```

```python
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color='green', marker='+', label='setosa')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color='blue', marker='.', label='versicolor')
```

```python
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'], color='green', marker='+', label='setosa')
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'], color='blue', marker='.', label='versicolor')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['target','flower_name'], axis='columns')
X.head()
```

```python
y = df.target
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
