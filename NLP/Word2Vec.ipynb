```python
import gensim
import pandas as pd
```

```python
df = pd.read_json("Cell_Phones_and_Accessories_5.json", lines=True)
df.head()
```

```python
df.shape
```

```python
df.reviewText[0]
```

```python
review_text=df.reviewText.apply(gensim.utils.simple_preprocess)
review_text
```

```python
model=gensim.models.Word2Vec(
    window=10, #take window of 10 words before the target
    min_count=2, #at least 2 words should be present in the text.
    workers=4 # how many CPU threads we want to use for the training.
)
```

```python
model.build_vocab(review_text, progress_per=1000)
```

```python
model.epochs
```

```python
model.train(review_text, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
model.save("./word2vec-amazon-cell-accessories-reviews-short.model")
```

```python
model.wv.most_similar("bad") #Gives the similar word which are close to "bad"
```

```python
model.wv.most_similar("chair")
```

```python
model.wv.similarity(w1="cheap", w2="inexpensive")  # cosine similarity score between 2 vectors
```

```python
model.wv.similarity(w1="good", w2="bad")
```
